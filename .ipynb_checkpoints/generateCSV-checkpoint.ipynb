# Conversão dos tweets de JSON para CSV

Colunas:
* Usuario
* Tweet
* Id
* Localização
* Hashtags
* Data/hora


```python
# encoding: utf-8
# encoding: iso-8859-1
# encoding: win-1252

import json
import csv

#https://coderwall.com/p/xdonta/how-to-convert-json-to-csv-excel

arquivo = open("data.lint.json")
tweets = json.load(arquivo)
arquivo.close()

with open("output.csv", 'w', encoding="utf-8") as  csvfile:
    spamwriter = csv.writer(csvfile, delimiter='|',quoting=csv.QUOTE_ALL)
    spamwriter.writerow(["usuario","tweet","id","localização","hashtags","data/hora"])
    for item in tweets:
        for node in tweets[item]:
            text = node["text"].replace("\n"," ")
            idTweet = node["id"]
            location = node["location"]
            hashtags = node["hashtags"]
            timestamp = node["timestamp"]
            linhaHashtags = ""
            if len(hashtags)>0:
                linhaHashtags = ""
                for hashtagIndividual in hashtags:
                    linhaHashtags += hashtagIndividual["text"] + ","
            spamwriter.writerow([item, idTweet, text, location, linhaHashtags, timestamp])
```
